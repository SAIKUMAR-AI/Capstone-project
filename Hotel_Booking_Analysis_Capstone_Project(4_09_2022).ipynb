# We have 3yrs of data - 2015-2017.
1. **Hotel**: Resort hotel or city hotel.
2. **Is_canceled**: 1 or 0 (cancelled or not cancelled).
3. **Lead_time**: Time of period for booking.
4. **arrival_date_month**: Customer arrived at the hotel in a month. 
5. country: Customer came from a country.
6. **Days_in_waiting_list**: After booking the time of delay getting confirmation.
7. **Deposite**: No-deposit, Non-Refundable, Refundable.
8. **Adr**: Average Daily Rate for an occupied room per day.
9. A**dults, Babies, children**: Number of Adults, Babies, children. 
10. **Assigned Room Type**: Type of room assigned to the customer.
11. **Booking Changes**: Number of changes.
12. **Distribution_channel**: Booking distribution channel.
13. **Is_repeated_guest**: Repeated guest(1) or not (0).
14. **Company**: ID of the company made the booking.
15. **Market_segment**: Market segment designation.
16. **Previous_cancellations**: Number of previous bookings that were cancelled by the customer prior to the current booking.
17. **Required_car_parking_spaces**: Number of car parking spaces required by the customer.
18. **Reservation_status**: Canceled – booking was canceled by the customer; Check Out – customer has checked in but already departed; No Show – customer did not check in and did inform the hotel of the reason why.
19. **Reservation_status_date**: Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel.
20. **Reserved_room_type**: Code of room type reserved

## **3.Data Visulization**

## **Percentage of Bookings in both Hotels.**

```python
#finding the percentage type of hotel where booked 
plt.pie(Hotel_data['hotel'].value_counts().tolist(),labels = Hotel_data['hotel'].value_counts().index.tolist(),autopct='%1.1f%%',startangle =90,textprops={'fontsize': 14})
plt.show()
```

```python
#number of hotels booked by customer in a month
month_hotel = Hotel_data[Hotel_data['is_canceled']==0].groupby(['arrival_date_month','hotel'])['hotel'].value_counts().unstack()
month_hotel.plot.bar(figsize=(16,12))
```
